Sample Exercise: Write a Python program to create a plot to get a general Statistics of the given dataset. Draw box plot, joinplot, scatterplot, pairplot, kernel density estimate plot (using seaborn) to explore the frequency of data in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/drive/MyDrive/mllab/dataset1.csv")
```

```python
df
```

```python
x=df['CRIM']
y=df['ZN']
plt.plot(x,y,'o',linestyle='-')
plt.xlabel('CRIM')
plt.ylabel('ZN')
plt.show()
```

```python
sns.jointplot(data=df,x='CRIM',y='ZN')
```

```python
plt.scatter(x,y)
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.kdeplot(df)
```

```python
plt.pie(df['AGE'])
plt.show()
```

```python
sns.heatmap(df)
```
